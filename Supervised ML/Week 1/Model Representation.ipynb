# Problem Statement
The following data about size vs price of houses is given:
| Size (1000 sqft)    | Price (1000s of dollars) |
| ------------------- | ------------------------ |
| 1.0                 | 300                      |
| 2.0                 | 500                      |

Find the price of a 1200 sqft house using a linear regression model.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train.shape = {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
i = 1
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker="x", color="red")
plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.show()
```

```python
w = 100
b = 200
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    
    for i in range(m):
        f_wb[i] = w * x[i] + b
   
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)
plt.plot(x_train, tmp_f_wb, color="blue", label="Our Prediction")
plt.scatter(x_train, y_train, marker="x", color="red", label="Actual Values")
plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.show()
```
